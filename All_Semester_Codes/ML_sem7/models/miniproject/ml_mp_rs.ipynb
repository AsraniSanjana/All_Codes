```python
import pandas as pd
import random

# Number of movies in the dataset
num_movies = 50

# Generate random data for movies
movie_data = {
    'movie_name': [f'Movie {i}' for i in range(1, num_movies + 1)],
    'genre': [random.choice(['Action', 'Drama', 'Comedy', 'Horror', 'Sci-Fi', 'Adventure', 'Romance', 'Fantasy', 'Animation', 'Mystery']) for _ in range(num_movies)],
    'suitable_for_age_grp': [random.choice(['Child', 'Teen', 'Adult', 'Senior']) for _ in range(num_movies)],
    'better_screened_on': [random.choice(['Independence Day', 'Halloween', 'Christmas', 'Valentine\'s Day', 'Other']) for _ in range(num_movies)],
    'suitable_for_gender': [random.choice(['Both', 'Female', 'Male']) for _ in range(num_movies)],
    'Movie_ID': [i for i in range(1, num_movies + 1)]  # Adding Movie IDs
}

# Create a DataFrame from the generated movie data
movie_data = pd.DataFrame(movie_data)

# Save the DataFrame to a CSV file
movie_data.to_csv('movie_data.csv', index=False)
```

```python
import pandas as pd
import random

# Load the Movie Data
movie_data = pd.read_csv('movie_data.csv')

# Load the People's Data
people_data = pd.read_csv('people_data.csv')

# Define the common special occasion for the day (e.g., Valentine's Day)
common_special_day = 'Valentine\'s Day'  # Change this to the actual special occasion

# Initialize a dictionary to store movie scores
movie_scores = {}

# Initialize counters for explanations
total_people = len(people_data)
female_count = len(people_data[people_data['Gender'] == 'Female'])
special_day_count = len(people_data[people_data['Special_Day'] == common_special_day])

# Loop through people's data
for index, person in people_data.iterrows():
    user_age = person['Age']
    user_gender = person['Gender']
    previous_movie_id = person['Movie_ID']

    # Filter Movies Based on User Characteristics and the Common Special Occasion
    filtered_movies = movie_data[
        (movie_data['suitable_for_age_grp'] == user_age) &
        (movie_data['better_screened_on'] == common_special_day) &
        ((movie_data['suitable_for_gender'] == 'Both') | (movie_data['suitable_for_gender'] == user_gender))
    ]

    if not filtered_movies.empty:
        # Profit Prediction Model (Scoring Model)
        filtered_movies = filtered_movies.copy()
        filtered_movies['profit_score'] = 0  # Initialize scores to 0

        # Assign scores based on genre, special day, and other factors
        # You can customize the scoring model based on your criteria.

        # Example: Assign higher scores for certain genres or special days
        filtered_movies.loc[filtered_movies['genre'] == 'Action', 'profit_score'] += 10
        filtered_movies.loc[filtered_movies['better_screened_on'] == common_special_day, 'profit_score'] += 5

        # Store recommended movies and their scores
        for index, row in filtered_movies.iterrows():
            movie_name = row['movie_name']
            score = row['profit_score']

            if movie_name in movie_scores:
                movie_scores[movie_name] += score
            else:
                movie_scores[movie_name] = score

# Aggregate the scores and recommend the movie with the highest score
recommended_movie = max(movie_scores, key=movie_scores.get)

print(f"Recommended Movie for Maximum Profit: {recommended_movie}")

# Provide an explanation for the recommendation
explanation = f"Out of {total_people} people present today, there are {female_count} females. " \
              f"Since the genre is {recommended_movie}, and today is {common_special_day}, screening " \
              f"{recommended_movie} will be profitable."

print("Explanation:")
print(explanation)
```
